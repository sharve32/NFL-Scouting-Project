```python
#Dependencies
import os
import csv
import pandas as pd
import numpy as np
from statistics import mean
```

```python
#Store filepaths
rb_import = "Output/RBs_Fltr_Combine_Data.csv"
rb_df = pd.read_csv(rb_import)
```

```python
list(rb_df)
```

```python
#Create lists for the combine columns
height = []
weight = []
hands = []
arm = []
wonderlic = []
forty = []
bench = []
vert = []
broad = []
shuttle = []
three_cone = []
sixty_shuttle = []

#Place columns into independent rows
for index, row in rb_df.iterrows():
    height.append(row['Height\xa0(in)'])
    weight.append(row['Weight\xa0(lbs)'])
    hands.append(row['Hand Size\xa0(in)'])
    arm.append(row['Arm Length\xa0(in)'])
    wonderlic.append([row['Wonderlic']])
    forty.append(row['40\xa0Yard'])
    bench.append(row['Bench Press'])
    vert.append(row['Vert Leap\xa0(in)'])
    broad.append(row['Broad Jump\xa0(in)'])
    shuttle.append(row['Shuttle'])
    three_cone.append(row['3Cone'])
    sixty_shuttle.append(row['60Yd Shuttle'])
    
```

```python
#Remove NaN values
hands = [x for x in hands if str(x) !='nan']
arm = [x for x in arm if str(x) !='nan']
wonderlic = [x for x in wonderlic if str(x) !='[nan]']
forty = [x for x in forty if str(x) !='nan']
bench = [x for x in bench if str(x) !='nan']
vert = [x for x in vert if str(x) !='nan']
broad = [x for x in broad if str(x) !='nan']
shuttle = [x for x in shuttle if str(x) !='nan']
three_cone = [x for x in three_cone if str(x) !='nan']
sixty_shuttle = [x for x in sixty_shuttle if str(x) !='nan']
```

```python
#Split lists into top 25 and bottom 25
#Wonderlic removed because not enough data
top_height = height[:25]
bottom_height = height[-25:]
top_weight = weight[:25]
bottom_weight = weight[-25:]
top_hands = hands[:25]
bottom_hands = hands[-25:]
top_arm = arm[:25]
bottom_arm = arm[-25:]
top_forty = forty[:25]
bottom_forty = forty[-25:]
top_bench = bench[:25]
bottom_bench = bench[-25:]
top_vert = vert[:25]
bottom_vert = vert[-25:]
top_broad = broad[:25]
bottom_broad = broad[-25:]
top_shuttle = shuttle[:25]
bottom_shuttle = shuttle[-25:]
top_three_cone = three_cone[:25]
bottom_three_cone = three_cone[-25:]
top_sixty_shuttle = sixty_shuttle[:25]
bottom_sixty_shuttle = sixty_shuttle[-25:]
```

```python
#Calculate average of each list
#Top 25
top_height_avg = mean(top_height)
top_weight_avg = mean(top_weight)
top_hands_avg = mean(top_hands)
top_arm_avg = mean(top_arm)
top_forty_avg = mean(top_forty)
top_bench_avg = mean(top_bench)
top_vert_avg = mean(top_vert)
top_broad_avg = mean(top_broad)
top_shuttle_avg = mean(top_shuttle)
top_three_cone_avg = mean(top_three_cone)
top_sixty_shuttle_avg = mean(top_sixty_shuttle)

#Bottom 25
bottom_height_avg = mean(bottom_height)
bottom_weight_avg = mean(bottom_weight)
bottom_hands_avg = mean(bottom_hands)
bottom_arm_avg = mean(bottom_arm)
bottom_forty_avg = mean(bottom_forty)
bottom_bench_avg = mean(bottom_bench)
bottom_vert_avg = mean(bottom_vert)
bottom_broad_avg = mean(bottom_broad)
bottom_shuttle_avg = mean(bottom_shuttle)
bottom_three_cone_avg = mean(bottom_three_cone)
bottom_sixty_shuttle_avg = mean(bottom_sixty_shuttle)

#Overall
height_avg = mean(height)
weight_avg = mean(weight)
hands_avg = mean(hands)
arm_avg = mean(arm)
forty_avg = mean(forty)
bench_avg = mean(bench)
vert_avg = mean(vert)
broad_avg = mean(broad)
shuttle_avg = mean(shuttle)
three_cone_avg = mean(three_cone)
sixty_shuttle_avg = mean(sixty_shuttle)
```

```python
#Create dictionary for column-oriented dataframe from the averages
# averages = {'Category': ['Top 25 Avg.', 'Bottom 25 Avg.'],
#            'Height (in)': [top_height_avg, bottom_height_avg],
#            'Weight (lbs)': [top_weight_avg, bottom_weight_avg],
#            'Hand Size (in)': [top_hands_avg, bottom_hands_avg],
#            'Arm Length (in)': [top_arm_avg, bottom_arm_avg],
#            '40 Time': [top_forty_avg, bottom_forty_avg],
#            'Bench Press': [top_bench_avg, bottom_bench_avg],
#            'Vertical Leap (in)': [top_vert_avg, bottom_vert_avg],
#            'Broad Jump (in)': [top_broad_avg, bottom_broad_avg],
#            'Shuttle': [top_shuttle_avg, bottom_shuttle_avg],
#            '3 Cone Drill': [top_three_cone_avg, bottom_three_cone_avg],
#            '60 Yard Shuttle': [top_sixty_shuttle_avg, bottom_sixty_shuttle_avg]}

averages = {'Category': ['Height (in)', 'Weight (lb)', 'Hand Size (in)', 'Arm Length (in)', '40 Yard', 'Bench Press',
                        'Vertical Leap (in)', 'Broad Jump (in)', 'Shuttle', '3 Cone Drill', '60 Yard Shuttle'],
            'Overall Avg.': [height_avg, weight_avg, hands_avg, arm_avg, forty_avg, bench_avg, vert_avg, broad_avg, shuttle_avg, 
                             three_cone_avg, sixty_shuttle_avg],
            'Top 25 Avg.': [top_height_avg, top_weight_avg, top_hands_avg, top_arm_avg, top_forty_avg, top_bench_avg,
                            top_vert_avg, top_broad_avg, top_shuttle_avg, top_three_cone_avg, top_sixty_shuttle_avg],
            'Bottom 25 Avg.': [bottom_height_avg, bottom_weight_avg, bottom_hands_avg, bottom_arm_avg, bottom_forty_avg, 
                               bottom_bench_avg, bottom_vert_avg, bottom_broad_avg, bottom_shuttle_avg, bottom_three_cone_avg,
                               bottom_sixty_shuttle_avg]}
```

```python
#Create dataframe from dictionary
averages_df = pd.DataFrame.from_dict(averages)
averages_df
```

```python
#Create function that calculates percent change
def pct_change(top, bottom):
    return (abs(top - bottom))/((top + bottom)/2) * 100
```

```python
#Add new column with percent change
averages_df['Pct. Change(%)'] = pct_change(averages_df['Top 25 Avg.'], averages_df['Bottom 25 Avg.'])
```

```python
averages_df
```

```python
#Calculate sum of pct. change column
total = averages_df['Pct. Change(%)'].sum()

#Created weighted percentage for each category, add to new column
averages_df['Weighted Pct.'] = (averages_df['Pct. Change(%)']/total)*100
```
