```python
#Dependencies
import os
import csv
import pandas as pd
import numpy as np
```

```python
#Store filepaths
combine_RB = "Output/RBs_Fltr_Combine_Data.csv"
combine_rb_df = pd.read_csv(combine_RB, encoding = "ISO-8859-1")
rb_avg = "Output/RBs_Averages.csv"
rb_avg_df = pd.read_csv(rb_avg)
```

```python
combine_rb_df.head()
```

```python
#Filter to only include 2015 drafted rbs
rbs_year = combine_rb_df [combine_rb_df['Year'] == 2015]
```

```python
rb_avg_df
```

```python
list(rbs_year)
```

```python
rbs_year.columns = ['Name', 'Pos', 'Year', 'Tm', 'First4AV', 'College', 'HOF', 'Class', 'Height (in)', 'Weight (lb)', 
                    'Hand Size (in)', 'Arm Length (in)', 'Wonderlic', '40 Yard', 'Bench Press', 'Vertical Leap (in)', 
                    'Broad Jump (in)', 'Shuttle', '3 Cone Drill', '60 Yard Shuttle']
del rbs_year['Wonderlic']
rbs_year.head()
```

```python
#Replace NaN with overall average for each category
rbs_year['Height (in)'].fillna(rb_avg_df['Overall Avg.'][0], inplace=True)
rbs_year['Weight (lb)'].fillna(rb_avg_df['Overall Avg.'][1], inplace=True)
rbs_year['Hand Size (in)'].fillna(rb_avg_df['Overall Avg.'][2], inplace=True)
rbs_year['Arm Length (in)'].fillna(rb_avg_df['Overall Avg.'][3], inplace=True)
rbs_year['40 Yard'].fillna(rb_avg_df['Overall Avg.'][4], inplace=True)
rbs_year['Bench Press'].fillna(rb_avg_df['Overall Avg.'][5], inplace=True)
rbs_year['Vertical Leap (in)'].fillna(rb_avg_df['Overall Avg.'][6], inplace=True)
rbs_year['Broad Jump (in)'].fillna(rb_avg_df['Overall Avg.'][7], inplace=True)
rbs_year['Shuttle'].fillna(rb_avg_df['Overall Avg.'][8], inplace=True)
rbs_year['3 Cone Drill'].fillna(rb_avg_df['Overall Avg.'][9], inplace=True)
rbs_year['60 Yard Shuttle'].fillna(rb_avg_df['Overall Avg.'][10], inplace=True)
```

```python
rbs_year.head()
```

```python
#Create function for weighted score
def weighted_score(ht, wt, hand, arm, forty, bench, vert, broad, shuttle, three_cone, sixty_yard):
    return .00122441*ht + .04237035*wt + .00005441*hand + .04285686*arm + .03818695*(6-forty) + .51948175*bench + .20683399*vert + .10368907*broad + .00731344*(6-shuttle) + .03164428*(9 - three_cone) + .00634449*(13 - sixty_yard)
```

```python
rbs_year['Weighted Score'] = weighted_score(rbs_year['Height (in)'], rbs_year['Weight (lb)'], rbs_year['Hand Size (in)'], 
                                            rbs_year['Arm Length (in)'], rbs_year['40 Yard'], rbs_year['Bench Press'], 
                                            rbs_year['Vertical Leap (in)'], rbs_year['Broad Jump (in)'], rbs_year['Shuttle'], 
                                            rbs_year['3 Cone Drill'], rbs_year['60 Yard Shuttle'])
```
