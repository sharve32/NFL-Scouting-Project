```python
#Dependencies
import os
import csv
import pandas as pd
import numpy as np
```

```python
#Store filepaths
draft_RB = "Resources/Draft_Results_1987-2015_RB.csv"
draft_rb_df = pd.read_csv(draft_RB)

draft_LB = "Resources/Draft_Results_1987-2015_LB.csv"
draft_lb_df = pd.read_csv(draft_LB)

combine_RB = "Resources/Combine_Data_1987-2019_RB.csv"
combine_rb_df = pd.read_csv(combine_RB, encoding = "ISO-8859-1")

combine_LB = "Resources/Combine_Data_1987-2019_LB.csv"
combine_lb_df = pd.read_csv(combine_LB, encoding = "ISO-8859-1")
```

```python
#Create boolean for whether player is in HOF or not
hof_rb = draft_rb_df['Player'].str.contains('HOF')
hof_lb = draft_lb_df['Player'].str.contains('HOF')

#Add new HOF column with values based on boolean
draft_rb_df['HOF'] = np.where(hof_rb, 'Y', 'N')
draft_lb_df['HOF'] = np.where(hof_lb, 'Y', 'N')

#Remove 'HOF' from player name
draft_rb_df['Player'] = draft_rb_df['Player'].map(lambda x: x.rstrip(' HOF'))
draft_lb_df['Player'] = draft_lb_df['Player'].map(lambda x: x.rstrip(' HOF'))
```

```python
#Rename column name Player to Name, and College/Univ to College
draft_rb_df = draft_rb_df.rename(index=str, columns={'Player':'Name', 'College/Univ':'College'})
draft_lb_df = draft_lb_df.rename(index=str, columns={'Player':'Name', 'College/Univ':'College'})
```

```python
#Create function to determine the class of each player
def player_class(c):
    if c['First4AV'] >= 50:
        return 'Top'
    elif c['First4AV'] < 20:
        return 'Bottom'
    else:
        return 'Mid'
```

```python
#Create new column "Class" to determine whether player is top, mid, or bottom
draft_rb_df['Class'] = draft_rb_df.apply(player_class, axis=1)
draft_lb_df['Class'] = draft_lb_df.apply(player_class, axis=1)
```

```python
#Merge draft and combine dfs on Name and College
rb_mrg_df = draft_rb_df.merge(combine_rb_df, how='inner', on=['Name', 'College'])
lb_mrg_df = draft_lb_df.merge(combine_lb_df, how='inner', on=['Name', 'College'])
list(rb_mrg_df)
```

```python
#Include only required columns
rb_fltr_df = rb_mrg_df.filter(['Name', 'Pos', 'Year_x', 'Tm', 'First4AV', 'College', 'HOF', 'Class', 'Height\xa0(in)',
                              'Weight\xa0(lbs)', 'Hand Size\xa0(in)', 'Arm Length\xa0(in)', 'Wonderlic', '40\xa0Yard',
                              'Bench Press', 'Vert Leap\xa0(in)', 'Broad Jump\xa0(in)', 'Shuttle', '3Cone', '60Yd Shuttle'], axis=1)
rb_fltr_df = rb_fltr_df.rename(index=str, columns={'Year_x':'Year'})
lb_fltr_df = lb_mrg_df.filter(['Name', 'Pos', 'Year_x', 'Tm', 'First4AV', 'College', 'HOF', 'Class', 'Height\xa0(in)',
                              'Weight\xa0(lbs)', 'Hand Size\xa0(in)', 'Arm Length\xa0(in)', 'Wonderlic', '40\xa0Yard',
                              'Bench Press', 'Vert Leap\xa0(in)', 'Broad Jump\xa0(in)', 'Shuttle', '3Cone', '60Yd Shuttle'], axis=1)
lb_fltr_df = lb_fltr_df.rename(index=str, columns={'Year_x':'Year'})
list(rb_fltr_df)
```
