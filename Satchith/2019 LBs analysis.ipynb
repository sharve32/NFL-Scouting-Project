```python
#Dependencies
import os
import csv
import pandas as pd
import numpy as np
```

```python
#Store filepaths
combine_LB = "Resources/Combine_Data_1987-2019_LB.csv"
combine_lb_df = pd.read_csv(combine_LB, encoding = "ISO-8859-1")
lb_avg = "Output/LBs_Averages.csv"
lb_avg_df = pd.read_csv(lb_avg)
```

```python
#Filter to only include 2019 drafted lbs
lbs_year = combine_lb_df [combine_lb_df['Year'] == 2019]
lbs_year
```

```python
lbs_year.columns = ['Year', 'Name', 'College', 'Pos', 'Height (in)', 'Weight (lb)', 'Hand Size (in)', 'Arm Length (in)', 
                    'Wonderlic', '40 Yard', 'Bench Press', 'Vertical Leap (in)', 'Broad Jump (in)', 'Shuttle', '3 Cone Drill', 
                    '60 Yard Shuttle']
del lbs_year['Wonderlic']
lbs_year.head()
```

```python
#Replace NaN with overall average for each category
lbs_year['Height (in)'].fillna(lb_avg_df['Overall Avg.'][0], inplace=True)
lbs_year['Weight (lb)'].fillna(lb_avg_df['Overall Avg.'][1], inplace=True)
lbs_year['Hand Size (in)'].fillna(lb_avg_df['Overall Avg.'][2], inplace=True)
lbs_year['Arm Length (in)'].fillna(lb_avg_df['Overall Avg.'][3], inplace=True)
lbs_year['40 Yard'].fillna(lb_avg_df['Overall Avg.'][4], inplace=True)
lbs_year['Bench Press'].fillna(lb_avg_df['Overall Avg.'][5], inplace=True)
lbs_year['Vertical Leap (in)'].fillna(lb_avg_df['Overall Avg.'][6], inplace=True)
lbs_year['Broad Jump (in)'].fillna(lb_avg_df['Overall Avg.'][7], inplace=True)
lbs_year['Shuttle'].fillna(lb_avg_df['Overall Avg.'][8], inplace=True)
lbs_year['3 Cone Drill'].fillna(lb_avg_df['Overall Avg.'][9], inplace=True)
lbs_year['60 Yard Shuttle'].fillna(lb_avg_df['Overall Avg.'][10], inplace=True)
```

```python
#Create function for weighted score
def weighted_score(ht, wt, hand, arm, forty, bench, vert, broad, shuttle, three_cone, sixty_yard):
    return .01202232*ht + .05327256*wt + .04130933*hand + .01830047*arm + .03818695*(6-forty) + .51948175*bench + 
           .20683399*vert + .10368907*broad + .00731344*(6-shuttle) + .03164428*(9 - three_cone) + .00634449*(13 - sixty_yard)
```

```python
lbs_year['Weighted Score'] = weighted_score(lbs_year['Height (in)'], lbs_year['Weight (lb)'], lbs_year['Hand Size (in)'], 
                                            lbs_year['Arm Length (in)'], lbs_year['40 Yard'], lbs_year['Bench Press'], 
                                            lbs_year['Vertical Leap (in)'], lbs_year['Broad Jump (in)'], lbs_year['Shuttle'], 
                                            lbs_year['3 Cone Drill'], lbs_year['60 Yard Shuttle'])
```

```python
highest_score = lbs_year.sort_values(by='Weighted Score', ascending=False)
highest_score
```

```python
lb_top_2019 = highest_score.filter(['Year', 'Name', 'College', 'Weighted Score'], axis=1)

lb_top_2019
```
